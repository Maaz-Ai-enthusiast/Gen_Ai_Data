```python
# it also tells whcih pdf page is blank

!pip install langchain langchain_community PyPDF2

import PyPDF2

# Step 1: Provide the file path (PDF file)
file_path = "/content/350KB.pdf"  # Replace with the correct path

# Step 2: Initialize a variable to store the extracted text
text = ""

try:
    # Step 3: Open the PDF file and read content
    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)

        # Step 4: Loop through all the pages and extract text
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            page_text = page.extract_text()

            # Check if page_text is not None (some PDFs may have pages without extractable text)
            if page_text:
                text += page_text
            else:
                print(f"Warning: No text found on page {page_num + 1}.")

except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
except Exception as e:
    print(f"An error occurred: {e}")

# Step 5: Print the entire extracted text
print(text)  # Print the entire text extracted from the PDF
```

**Web page URL code**

```python
# Install the necessary libraries
!pip install langchain langchain_community beautifulsoup4 requests

# Import the necessary modules
import requests
from bs4 import BeautifulSoup

# Step 1: Specify the URL of the webpage you want to load
url = "https://csoftsystems.com/"  # Replace with the desired URL

# Step 2: Set up headers to avoid being blocked
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    # Step 3: Fetch the HTML content from the URL
    response = requests.get(url, headers=headers)

    # Check if the request was successful
    response.raise_for_status()  # Raises an error for bad responses (4XX or 5XX)

    # Step 4: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Step 5: Extract text from the parsed HTML
    text = soup.get_text(separator="\n")  # You can adjust the separator as needed

    # Step 6: Print the extracted text (formatted)
    print("Extracted Text:\n")
    print(text.strip())  # Print the entire text, trimmed of leading/trailing whitespace

except requests.exceptions.HTTPError as http_err:
    print(f"HTTP error occurred: {http_err}")
except requests.exceptions.RequestException as err:
    print(f"Error occurred: {err}")
```

1. PyPDFLoader
Description: Utilizes the pypdf library to read and extract text from PDF files.

Pros:

Lightweight: Simple to implement and use for basic text extraction tasks.
Efficiency: Fast for standard PDFs that contain primarily text.
Cons:

Limited Capability: Struggles with complex PDF layouts, such as those with multiple columns, images, or intricate formatting.
No Support for Encrypted PDFs: Cannot extract text from password-protected PDF files.
2. PDFPlumberLoader
Description: Leverages the pdfplumber library, which provides advanced capabilities for extracting text and data.

Pros:

Robust Handling: Excels at extracting text from complex layouts, including those with tables, images, and varied formatting.
Structured Data Extraction: Can extract tables and other structured data directly, making it ideal for data analysis tasks.
Cons:

Performance: Slightly slower than simpler loaders due to its comprehensive processing.
Complexity: May require additional handling and code to extract structured data effectively.
3. PyMuPDFLoader
Description: Utilizes the PyMuPDF (also known as fitz) library, which is optimized for high-performance PDF processing.

Pros:

Speed: Fast extraction of both text and images from PDFs.
Versatile: Capable of handling a wide variety of PDF elements, including annotations and graphical content, making it suitable for multimedia PDFs.
Cons:

Setup Complexity: More complex to install and configure compared to simpler options like PyPDFLoader.
Parameter Handling: May require additional parameters and customization to extract specific content types.
Summary
Choose PyPDFLoader for straightforward text extraction from standard PDF documents.
Opt for PDFPlumberLoader when working with complex documents that require data extraction from tables or intricate layouts.
Select PyMuPDFLoader for high-performance extraction needs, particularly when dealing with graphics or multimedia elements in PDFs.
