```python
# Install the necessary libraries
!pip install openai PyMuPDF

import openai
import fitz  # PyMuPDF
from google.colab import files

# Set your OpenAI API key
openai.api_key = userdata.get('Pdfchatbot') # Replace with your actual API key

# Load and extract text from PDF
def load_pdf_text(file_path):
    pdf_document = fitz.open(file_path)
    text = ""

    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        text += page.get_text()  # Concatenate text from all pages

    pdf_document.close()
    return text

# Check for the existence of a line or topic
def check_line_existence(text, query):
    return query.lower() in text.lower()

# Generate MCQs using the OpenAI API
def generate_mcqs_with_llm(source_text, topic, num_mcqs):
    prompt = f"Generate {num_mcqs} multiple choice questions about '{topic}' based on the following text:\n\n{source_text}\n\nFormat your response with each question followed by four options and indicate the correct answer."

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",  # or your preferred model
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    return response['choices'][0]['message']['content']

# Chatbot Functionality
def chatbot():
    # Step 1: Upload the PDF file
    uploaded = files.upload()

    for file_name in uploaded.keys():
        pdf_text = load_pdf_text(file_name)

    while True:
        print("\nOptions:")
        print("1. Are you looking for a certain line/topic?")
        print("2. Do you want to test yourself with certain topics?")
        print("3. Exit")

        choice = input("Select an option (1/2/3): ")

        if choice == '1':
            query = input("Enter the line/topic you're looking for: ")
            exists = check_line_existence(pdf_text, query)
            if exists:
                print(f"The topic '{query}' exists in the PDF.")
            else:
                print(f"The line/topic '{query}' does not exist in the PDF.")

        elif choice == '2':
            topic = input("Enter the topic you want to test yourself on: ")
            exists = check_line_existence(pdf_text, topic)
            if exists:
                num_mcqs = int(input("How many MCQs do you want to generate? "))
                mcqs = generate_mcqs_with_llm(pdf_text, topic, num_mcqs)
                print("\nGenerated MCQs:")
                print(mcqs)  # Display the generated MCQs
            else:
                print(f"The topic '{topic}' does not exist in the PDF.")

        elif choice == '3':
            print("Exiting the chatbot.")
            break

        else:
            print("Invalid choice, please try again.")

# Run the chatbot
chatbot()
```
